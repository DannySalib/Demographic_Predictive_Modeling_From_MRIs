creating a client

```python
import boto3
from botocore import UNSIGNED
from botocore.client import Config

# Set up an anonymous client
s3 = boto3.client('s3', 
                  endpoint_url='https://s3.amazonaws.com',
                  config=Config(signature_version=UNSIGNED))  # Disable signing
```

```python
# List all files in the dataset
paginator = s3.get_paginator('list_objects_v2')
for page in paginator.paginate(Bucket='openneuro.org', Prefix='ds000228/'):
    for obj in page.get('Contents', []):
        print(obj['Key'])  # Print all file paths
```

```python
# Downloading file locally
key = 'ds000228/sub-pixar001/anat/sub-pixar001_T1w.nii.gz'
file_name = 'sub-01_T1w.nii.gz' 

s3.download_file(
    Bucket='openneuro.org',
    Key=key,  # Use the exact path from list_objects_v2
    Filename= file_name
)

import os
assert os.path.exists(file_name)
print('Success!')
```
